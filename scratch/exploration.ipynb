```python
import pandas as pd
from PIL import Image
import torch
from torch.nn.functional import cosine_similarity
from transformers import CLIPProcessor, CLIPModel
```

```python
model = CLIPModel.from_pretrained("openai/clip-vit-large-patch14")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-large-patch14")
```

```python
def compute_similarity(image_path, text):
  image = Image.open(image_path)
  inputs = processor(text=[text], images=image, return_tensors="pt", padding=True, truncation=True)
  outputs = model(**inputs)
  
  # Get the image and text features from the model
  image_features = outputs.image_embeds  # Image features
  text_features = outputs.text_embeds    # Text features

  # Compute cosine similarity as the similarity score
  similarity_score = cosine_similarity(image_features, text_features).item()
  return similarity_score
```

```python
# Read the csv
df = pd.read_csv('./challenge_set/challenge_set.csv')

def format_url_to_id(url):
  parts = url.split('/')
  user_id = parts[4]  # Based on the structure of your URL
  return f"./challenge_set/{user_id}.png"

# Apply the function to each URL in the dataframe
df['formatted'] = df['url'].apply(format_url_to_id)
```

```python
df['similarity_score'] = df.apply(lambda row: compute_similarity(row['formatted'], row['caption']), axis=1)
```

```python
df.sort_values(by='similarity_score', ascending=False, inplace=True)
df
```
